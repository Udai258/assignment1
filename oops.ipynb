13) What is a property decorator in Python?
Ans:)A property decorator in Python is used to define getter, setter, and deleter methods for class attributes in an elegant way. Here are five key points:

@property Decorator – Converts a method into a getter, allowing access to a private attribute as if it were a property.
@<property_name>.setter – Defines a setter method to modify the property while enforcing validation or constraints.
@<property_name>.deleter – Allows defining a deleter method to delete the property.
Encapsulation – Provides controlled access to private attributes while maintaining an intuitive syntax.
Example Usage – Helps create read-only properties or properties with controlled modification behavior.
Example:

class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name  # Getter

    @name.setter
    def name(self, new_name):
        if isinstance(new_name, str):
            self._name = new_name  # Setter
        else:
            raise ValueError("Name must be a string")

    @name.deleter
    def name(self):
        del self._name  # Deleter

p = Person("Alice")
print(p.name)  # Calls getter
p.name = "Bob"  # Calls setter
del p.name  # Calls deleter

17) What is the difference between a class variable and an instance variable?
Ans:)The main differences between class variables and instance variables in Python are:

Definition & Scope:

Class Variable: Shared across all instances of a class and defined outside any method using the class name.
Instance Variable: Unique to each instance and defined inside a method using self.
Memory Allocation:

Class Variable: Stored in a shared memory location for all objects of the class.
Instance Variable: Stored separately for each instance.
Access & Modification:

Class Variable: Can be accessed using ClassName.variable_name or self.variable_name. Changes affect all instances if modified at the class level.
Instance Variable: Accessed and modified using self.variable_name. Changes affect only that specific instance.
Usage:

Class Variable: Used for constants or shared properties (e.g., keeping track of object count).
Instance Variable: Stores unique object-specific data.
Example:


class Car:
    wheels = 4  # Class variable (shared)

    def __init__(self, color):
        self.color = color  # Instance variable (unique per object)

car1 = Car("Red")
car2 = Car("Blue")

print(car1.wheels, car1.color)  # Output: 4, Red
print(car2.wheels, car2.color)  # Output: 4, Blue

Car.wheels = 6  # Modifying class variable affects all instances
print(car1.wheels, car2.wheels)  # Output: 6, 6

18) What is multiple inheritance in Python?
Ans:)Multiple inheritance in Python allows a class to inherit from more than one parent class. Here are five key points:

Definition – A subclass can inherit attributes and methods from multiple parent classes, combining their functionality.
Syntax – Defined by specifying multiple parent classes in parentheses during subclass declaration.
Method Resolution Order (MRO) – Python follows the C3 Linearization (MRO) to determine the order in which methods are inherited, using the super() function.
Code Reusability – Allows a subclass to reuse code from multiple base classes, reducing redundancy.
Potential Complexity – Can lead to conflicts if different parent classes have methods with the same name, requiring careful MRO handling.
Example:

class A:
    def show(self):
        print("Class A")

class B:
    def display(self):
        print("Class B")

class C(A, B):  # Multiple Inheritance
    pass

obj = C()
obj.show()    # Output: Class A
obj.display() # Output: Class B
Here, C inherits methods from both A and B.

 Ans:)The @staticmethod and @classmethod decorators in Python modify method behavior in a class. Here are five key differences:

Use Case:

@staticmethod does not depend on class inheritance.

python
